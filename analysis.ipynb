```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
balance = pd.read_csv("df_balance.csv")
flujo = pd.read_csv("df_flujo.csv")
resultados = pd.read_csv("df_resultados.csv")
```

```python
df = pd.concat([balance.drop(balance.columns[0], axis=1), flujo.drop(flujo.columns[0], axis=1), resultados.drop(resultados.columns[0], axis=1)], axis=0)
```

```python
df.pivot(index=["year", "trimestre"], columns = "cuentas", values="valor")
df[df.columns[0:5]] = df[df.columns[0:5]].apply(lambda x: x.str.replace(',', '')).astype(float)
df['Liquidez Corriente'] = df['Total Activos Corrientes']/df['Total Pasivos Corrientes']
df['Solvencia'] = (df['Total Pasivos Corrientes'] + df['Total Pasivos No Corrientes'])/df['Total Patrimonio']
df['Endeudamiento Total'] = (df['Total Pasivos Corrientes'] + df['Total Pasivos No Corrientes'])/(df['Total Activos Corrientes'] + df['Total Activos No Corrientes'])
df['']

#
df = df.reset_index()

# Create a mapping from Roman numerals to integers for quarters
quarter_map = {
    'I': 1,
    'II': 2,
    'III': 3,
    'IV': 4
}

# Map the quarters to numeric format
df['trimestre'] = df['trimestre'].map(quarter_map)

df['time'] = df['year'].astype('str') + '-Q' + df['trimestre'].astype('str')

plt.figure(figsize=(8, 6))
```

```python
def plot_balance():
    '''
    Plot financial indicators derived from Balance General
    '''
    balance = pd.read_csv("df_balance.csv")
    balance = balance.drop(balance.columns[0], axis=1)
    balance = balance.pivot(index=["year", "trimestre"], columns = "cuentas", values="valor")
    balance[balance.columns[0:5]] = balance[balance.columns[0:5]].apply(lambda x: x.str.replace(',', '')).astype(float)
    balance['Liquidez Corriente'] = balance['Total Activos Corrientes']/balance['Total Pasivos Corrientes']
    balance['Solvencia'] = (balance['Total Pasivos Corrientes'] + balance['Total Pasivos No Corrientes'])/balance['Total Patrimonio']
    balance['Endeudamiento Total'] = (balance['Total Pasivos Corrientes'] + balance['Total Pasivos No Corrientes'])/(balance['Total Activos Corrientes'] + balance['Total Activos No Corrientes'])
    balance = balance.reset_index()

    # Create a mapping from Roman numerals to integers for quarters
    quarter_map = {
        'I': 1,
        'II': 2,
        'III': 3,
        'IV': 4
    }

    # Map the quarters to numeric format
    balance['trimestre'] = balance['trimestre'].map(quarter_map)

    balance['time'] = balance['year'].astype('str') + '-Q' + balance['trimestre'].astype('str')


    plt.figure(figsize=(8, 6))

    # Plot the financial indicators over time
    plt.plot(balance['time'], balance['Liquidez Corriente'], marker='o', label='Liquidez Corriente')
    plt.plot(balance['time'], balance['Solvencia'], marker='s', label='Solvencia')
    plt.plot(balance['time'], balance['Endeudamiento Total'], marker='^', label='Endeudamiento Total')

    # Customize the plot
    plt.xlabel('')
    plt.ylabel('')
    plt.title('Evoluci√≥n de Indicadores Financieros - Balance General')
    plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
    plt.legend()

    # Display the plot
    plt.tight_layout()
    plt.show()
```

```python
plot_balance()
```
